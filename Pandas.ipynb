# High level data understanding.

Bivariate analysis:

Between two columns.
exmaple: eat more= gain more weight.

crosstable=> crosstab()

```python
pd.crosstab(df1["Gender"],df1["Loan_Status"])
```

```python
pd.crosstab(df1["Gender"],df1["Loan_Status"],margins=True)
```

```python
pd.crosstab(df1["Gender"],df1["Loan_Status"],margins=True,normalize="index")*100
```

```python
pd.crosstab(df1["Self_Employed"],df1["Loan_Status"]).plot(kind="bar")
```

```python
pd.crosstab(df1["Self_Employed"],df1["Loan_Status"])
```

```python
pd.crosstab(df1["Credit_History"],df1["Loan_Status"])
```

```python
pd.crosstab(df1["Credit_History"],df1["Loan_Status"]).plot(kind="bar")
```
