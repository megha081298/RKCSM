```python
np.random.seed(6)
Matrix=np.random.randint(10,101,(5,4)) #(rows,cols)

#top 2 rows with all the columns
print(Matrix)
Matrix[0:2,:]
```

```python
#All the rows and 1st two cols:

Matrix[:,0:2]
```

```python
#1st 2 rows and 1st 2 cols:

Matrix[0:2,0:2]
```

```python
#2nd and 5th row and 3 and 4th column


Matrix[[1,4],2:4]
```
