# Broadcasting

1.Broadcasting allows you to perform arithematic operations between arrays with different shapes or sizes.

2.Broadcasting essentially streches the smaller ndarray so that it matches the shape of larger ndarray.

3.Limitations of python list: python list with different sizes cannot be added subtracted or perform any arithematic operations.

```python
np.array([10,20,30])+np.array([5,5,5])
```

```python
v=np.array([10,20,30])+np.array([5])
print(v)
```
